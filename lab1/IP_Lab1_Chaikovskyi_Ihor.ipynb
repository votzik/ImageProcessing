```python
import numpy as np
def linear_filter(img, K, L): 
  img = np.asarray(img, dtype=np.float64) 
  img = img*K + L 
  # clip pixel values 
  img[img > 255] = 255 
  img[img < 0] = 0 
  return np.asarray(img, dtype = np.int64) 
```

```python
import matplotlib.pyplot as plt
def show_images(im_list):
  fig = plt.figure(figsize=(24,12))
  result = np.hstack(im_list) 
  plt.imshow(result) 
  plt.axis('off') 
  plt.show() 
```

```python
import matplotlib.image as mpimg
low_res = mpimg.imread('/content/low-res.jpg')
high_res = mpimg.imread('/content/high-res.jpg')
low_con = mpimg.imread('/content/low-contrast.jpg')
high_con = mpimg.imread('/content/high-contrast.jpg')
```

```python
K_list = [0.5, 1, 0.8, 1]
L_list = [0, 20, 40, -40]
```

```python
low_res_result = [low_res]
for K, L in zip(K_list, L_list):
  result_im = linear_filter(low_res, K, L)
  low_res_result.append(result_im)

show_images(low_res_result)
```

```python
high_res_result = [high_res]
for K, L in zip(K_list, L_list):
  result_im = linear_filter(high_res, K, L)
  high_res_result.append(result_im)

show_images(high_res_result)
```

```python
low_con_result = [low_con]
for K, L in zip(K_list, L_list):
  result_im = linear_filter(low_con, K, L)
  low_con_result.append(result_im)

show_images(low_con_result)
```

```python
high_con_result = [high_con]
for K, L in zip(K_list, L_list):
  result_im = linear_filter(high_con, K, L)
  high_con_result.append(result_im)

show_images(high_con_result)
```
